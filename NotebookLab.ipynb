### Task 1

We typeset the following logarithm properties from Appendix A of Bramer:

$$\log_{2}(a \times b) = \log_{2} a + \log_{2} b$$
$$\log_{2}(a / b) = \log_{2} a - \log_{2} b$$
$$\log_{2}(a^{n}) = n \times \log_{2} a$$
$$\log_{2}(1 / a) = - \log_{2} a$$

We also typeset the summation property:

$$\sum_{i=1}^{i=m} \sum_{j=1}^{j=n} a_{ij} = \sum_{j=1}^{j=n} \sum_{i=1}^{i=m} a_{ij}$$

Source: Bramer, Max. *Principles of Data Mining*

The formula for the mean of a sequence of numbers $a_1, a_2, \ldots, a_n$ is:

$$\bar{a} = \frac{\sum_{i=1}^{n} a_i}{n}$$

```python
import random

# function to compute the mean of a list
def mean(numbers):
  return sum(numbers) / len(numbers) if numbers else float('nan')

# generate two lists with 100 random numbers each (0-100)
list1 = [random.uniform(0, 100)]
list2 = [random.uniform(0, 100)]

# calculate and display means
print("Mean of list 1:", mean(list1))
print("Mean of list 2:", mean(list2))
```
